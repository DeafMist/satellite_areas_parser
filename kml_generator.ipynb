```python
from lxml import etree
from pykml.parser import Schema
from pykml.factory import KML_ElementMaker as KML


def kml_generator(name, values, coordinates):
  # Параметр name - Имя спутника_Название луча
  doc = KML.kml(
      KML.Document(
          KML.name(name + '.kml'),
          KML.Folder(
              KML.name(name)
          )
      )
  )

  polygons_number = len(values)

  # идем по количеству полигонов
  # Параметр values[i] - значение EIRP, coordinates - пары координат lon,lat, разделенные пробелами  
  for i in range(1, polygons_number + 1):
    doc.Document.Folder.append(
      KML.Placemark(
        KML.name(i),
        KML.ExtendedData(
          KML.Data(
            KML.value(values[i - 1]),
            name='EIRP'
          )
        ),
        KML.Polygon(
          KML.tesselate('1'),
          KML.extrude('1'),
          KML.altitudeMode('clampedToGround'),
          KML.outerBoundaryIs(
            KML.LinearRing(
              KML.coordinates(
                coordinates[i - 1]
              )
            )
          )
        )   
      )  
    ) 

  # check that the KML document is valid using the Google Extension XML Schema
  # assert(Schema("kml22gx.xsd").validate(doc))

  # print(etree.tostring(doc, pretty_print=True))

  with open(name + '.kml', 'wb') as f:
    f.write(etree.tostring(doc, pretty_print=True))
```
